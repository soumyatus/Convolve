```python
def add_one_to_slots(slot_list):
    # Convert the string to a list of integers, add 1 to each, and join back into a string
    return ' '.join(str(int(slot) + 1) for slot in slot_list.split())

# Apply the function to the 'predicted_slots' column
final_result['predicted_slots_order'] = final_result['predicted_slots_order'].apply(add_one_to_slots)

# Display the updated DataFrame
print(final_result)
```

```python
# Check if NaN values remain in numeric columns
print(df[numeric_columns].isna().sum())
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['product_category', 'product_sub_category', 'Offer_id','Offer_subid'])

# Drop 'customer_code' if present
if 'customer_code' in df.columns:
    df = df.drop(columns=['customer_code'])

# Adjust 'slot' column to start from 0
df['slot'] = df['slot'] - 1
df['slot'] = df['slot'].astype(int)

# Create features and targets
X = df.drop(columns=['flag', 'slot'])  # Features
y = df['slot']  # Target (slots 0-27)
weights = df['flag']  # Use 'flag' as sample weights

# Split the data into training and testing sets
X_train, X_test, y_train, y_test, weights_train, weights_test = train_test_split(
    X, y, weights, test_size=0.2, random_state=42
)

# Initialize the XGBoost model
xgb_model = xgb.XGBClassifier(
    n_estimators=100,        # Number of boosting rounds
    max_depth=6,             # Maximum tree depth
    learning_rate=0.1,       # Learning rate
    objective='multi:softprob',  # Multi-class classification with probabilities
    num_class=28,            # Number of classes (slots 0 to 27)
    eval_metric='mlogloss',  # Multi-class log loss for evaluation
    use_label_encoder=False  # Avoid warning about label encoding
)

# Train the model
xgb_model.fit(
    X_train,
    y_train,
    sample_weight=weights_train,  # Use 'flag' as sample weights
    eval_set=[(X_test, y_test)],
    verbose=True
)

# Predict on the test set
y_pred = xgb_model.predict(X_test)
y_pred_proba = xgb_model.predict_proba(X_test)

# Evaluate the accuracy of the predictions
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")

# Output probabilities sorted by priority for each customer
results = []
for i, customer in enumerate(X_test.index):
    probs = y_pred_proba[i]
    sorted_slots = np.argsort(-probs)  # Sort slots by descending probability
    result = {
        'customer': customer,
        'sorted_slots': sorted_slots.tolist(),
        'probabilities': probs[sorted_slots].tolist()
    }
    results.append(result)

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame(results)
print(results_df.head())  # Display the top results
```
