```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['product_category', 'product_sub_category'])

# Drop 'customer_code' if present
if 'customer_code' in df.columns:
    df = df.drop(columns=['customer_code'])

# Adjust 'slot' column to start from 0
df['slot'] = df['slot'] - 1
df['slot'] = df['slot'].astype(int)

# Create features and targets
X = df.drop(columns=['flag', 'slot'])  # Features
y = df['slot']  # Target (slots 0-27)
weights = df['flag']  # Use 'flag' as sample weights

# Split the data into training and testing sets
X_train, X_test, y_train, y_test, weights_train, weights_test = train_test_split(
    X, y, weights, test_size=0.2, random_state=42
)

# Initialize the XGBoost model
xgb_model = xgb.XGBClassifier(
    n_estimators=100,        # Number of boosting rounds
    max_depth=6,             # Maximum tree depth
    learning_rate=0.1,       # Learning rate
    objective='multi:softprob',  # Multi-class classification with probabilities
    num_class=28,            # Number of classes (slots 0 to 27)
    eval_metric='mlogloss',  # Multi-class log loss for evaluation
    use_label_encoder=False  # Avoid warning about label encoding
)

# Train the model
xgb_model.fit(
    X_train,
    y_train,
    sample_weight=weights_train,  # Use 'flag' as sample weights
    eval_set=[(X_test, y_test)],
    verbose=True
)

# Predict on the test set
y_pred = xgb_model.predict(X_test)
y_pred_proba = xgb_model.predict_proba(X_test)

# Evaluate the accuracy of the predictions
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")

# Output probabilities sorted by priority for each customer
results = []
for i, customer in enumerate(X_test.index):
    probs = y_pred_proba[i]
    sorted_slots = np.argsort(-probs)  # Sort slots by descending probability
    result = {
        'customer': customer,
        'sorted_slots': sorted_slots.tolist(),
        'probabilities': probs[sorted_slots].tolist()
    }
    results.append(result)

# Convert results to a DataFrame for better visualization
results_df = pd.DataFrame(results)
print(results_df.head())  # Display the top results
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb

# Load the test dataset
df_test = pd.read_csv('/content/drive/MyDrive/data/TESTFINALDATA.csv')

# Load the submission order file (contains the required `customer_code` order)
submission_order = pd.read_csv('/content/drive/MyDrive/data/Test/test_customers.csv')

# Load the trained model
# xgb_model = xgb.XGBClassifier()
xgb_model.load_model('/content/drive/MyDrive/data/xgb_model.json')  # Replace with your saved model path

# Preprocessing function for test data
def preprocess_test_data(df, training_columns):
    # Create 'slot' column if not present
    if 'open_slot' in df.columns and 'send_slot' in df.columns:
        df['slot'] = df['open_slot'].fillna(df['send_slot'])

    # Drop irrelevant columns
    columns_to_drop = ['send_timestamp', 'open_timestamp', 'original_index', 'send_slot', 'open_slot']
    df = df.drop(columns=columns_to_drop, errors='ignore')

    # One-hot encode categorical columns
    df = pd.get_dummies(df, columns=['product_category', 'product_sub_category'])

    # Align columns with the training data
    df = df.reindex(columns=training_columns, fill_value=0)

    return df

# Extract the training feature names from the model
training_columns = xgb_model.get_booster().feature_names

# Sort the test dataset based on the order of `customer_code` in `submission_order`
df_test_sorted = df_test.set_index('customer_code').reindex(submission_order['customer_code']).reset_index()

# Preprocess the test data
X_test = preprocess_test_data(df_test_sorted, training_columns)

# Predict probabilities for each slot
y_pred_proba = xgb_model.predict_proba(X_test)

# Generate the slot order for each customer
results = []
for i, customer_code in enumerate(df_test_sorted['customer_code']):
    probs = y_pred_proba[i]
    sorted_slots = np.argsort(-probs)  # Sort slots by descending probability
    result = {
        'customer_code': customer_code,
        'order': " ".join(map(str, sorted_slots))  # Convert slots to a space-separated string
    }
    results.append(result)

# Create a DataFrame from results
results_df = pd.DataFrame(results)

# # Ensure the submission order matches the test_customers.csv order
# submission_df = submission_order.merge(results_df, on='customer_code', how='left')

# # Save the submission file
# submission_file_path = '/content/drive/MyDrive/data/Test/test_submission.csv'
# submission_df.to_csv(submission_file_path, index=False)

# print(f"Submission file saved to: {submission_file_path}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBClassifier
from tqdm import tqdm

# Step 1: Prepare Data
print("Preparing data for training...")

# Copy the merged data to a new variable
training_data = merged_df.copy()

# Feature Engineering
training_data['hour_of_send'] = pd.to_datetime(training_data['send_timestamp']).dt.hour
training_data['day_of_week'] = pd.to_datetime(training_data['send_timestamp']).dt.weekday
training_data['is_weekend'] = training_data['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Drop unnecessary columns including 'open_slot'
columns_to_drop = ['original_index', 'send_timestamp', 'open_timestamp', 'flag', 'open_slot']
training_data_cleaned = training_data.drop(columns=columns_to_drop)

# Encode categorical variables
print("Encoding categorical variables...")
encoder_updated = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
categorical_encoded_updated = encoder_updated.fit_transform(
    training_data_cleaned[['product_category', 'product_sub_category']]
)

# Combine numerical and categorical features
numerical_features_updated = training_data_cleaned.select_dtypes(include=['float64', 'int64'])
X_updated = np.hstack((numerical_features_updated, categorical_encoded_updated))
y_updated = training_data_cleaned['send_slot'] - 1  # Target variable (zero-indexed)

# Filter underrepresented classes
print("Filtering underrepresented classes...")
class_counts_updated = pd.Series(y_updated).value_counts()
valid_classes_updated = class_counts_updated[class_counts_updated >= 2].index
filtered_indices_updated = y_updated.isin(valid_classes_updated)
X_filtered_updated = X_updated[filtered_indices_updated]
y_filtered_updated = y_updated[filtered_indices_updated]

# Train-Test Split
print("Splitting data...")
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(
    X_filtered_updated, y_filtered_updated, test_size=0.2, random_state=42, stratify=y_filtered_updated
)

# Step 2: Train the Model
print("Training the XGBoost model...")
slot_prediction_model_final = XGBClassifier(
    objective='rank:pairwise',
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    eval_metric='mlogloss',
    random_state=42
)

# Use a progress bar for monitoring training
with tqdm(total=1, desc="Training Progress", unit="iterations") as progress_bar:
    slot_prediction_model_final.fit(
        X_train_final,
        y_train_final,
        eval_set=[(X_test_final, y_test_final)],
        verbose=True
    )
    progress_bar.update(1)

# Step 3: Evaluate the Model
print("\nEvaluating the model...")
y_pred_final = slot_prediction_model_final.predict(X_test_final)
accuracy_final = (y_pred_final == y_test_final).mean()
print(f"\nAccuracy Score: {accuracy_final:.4f}")

# Step 4: Generate Slot Rankings
print("\nGenerating slot rankings...")
slot_probabilities_final = slot_prediction_model_final.predict_proba(X_test_final)
ranked_slots_final = np.argsort(-slot_probabilities_final, axis=1) + 1  # Ranking slots based on probabilities

# Combine Predictions with Customer Codes
results_final = pd.DataFrame({
    'customer_code': training_data_cleaned.iloc[y_test_final.index]['customer_code'].values,
    'predicted_slots_order': [list(row) for row in ranked_slots_final],
})

# Save Predictions
results_final.to_csv('predicted_slots_final.csv', index=False)
print("\nPredicted slot order saved as 'predicted_slots_final.csv'.")

# Display a sample of the results
print("\nSample Predictions:")
print(results_final.head())
```
